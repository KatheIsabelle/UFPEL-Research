```python
# IMPLEMENTAÇÃO DE GRAFOS TASK 2 (09/04)

import networkx as nx
import matplotlib.pyplot as plt


def cria_grafo():
    G = nx.Graph()                #cria um grafo direcionado  
    return G


def add_nó(G):
    G.add_nodes_from([6,7,8,9,10,11,12,13]) #adiciona mais de um nó ao grafo
    print ("OS NÓS INICIAIS SÃO:", G.nodes())
  
  
def add_arestas(G):
    G.add_edges_from([(6,7),(7,8),(8,9),(9,10),(10,11),(11,12),(12,13),(13,6)]) #adiciona uma aresta ao grafo
    print ("AS ARESTAS INICIAS SÃO:", G.edges())
    print("\n\nRESUMO ANTES DA MODIFICAÇÃO\n","Nós:", G.number_of_nodes(), "Arestas:", G.number_of_edges())


def exclui_nó(G):
    exclui = int (input ("Digite o nó que deseja excluir: "))
    
    if G.has_node(exclui):
        
        G.remove_node(exclui)
        
    else:
        print("\nO NÓ ESCOLHIDO PARA SER REMOVIDO NÃO EXISTE!")
    
 
def remove_aresta(G):
    edge_to_remove = input("Digite a aresta que deseja excluir (no formato 'nó1 nó2'): ")
    node1, node2 = edge_to_remove.split()
    if G.has_edge(int(node1), int(node2)):
        G.remove_edge(int(node1), int(node2))
    else:
        print("A aresta não existe no grafo.")
    
 
  
def print_grafo(G):
    print("_________________________________________________________")
    print("\n OS NÓS PÓS-REMOVER:")
    print(G.nodes())
    print("\n AS ARESTAS PÓS-REMOVER:")
    print(G.edges())
    print("\n\nRESUMO DEPOIS DA MODIFICAÇÃO\n","Nós:", G.number_of_nodes(), "Arestas:", G.number_of_edges())
    
    
def draw_grafo(G): 
  nx.draw(G, with_labels=True, font_weight='bold')
  plt.show(G)
    
    
#MAIN_______________________________________________________________
G = cria_grafo()
add_nó(G)
add_arestas(G)
exclui_nó(G)
remove_aresta(G)
print_grafo(G)
draw_grafo(G)
```
